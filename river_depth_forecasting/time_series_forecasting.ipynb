filling for missing values in hydrometry and drainage_volume features

options include: 

1. fill with zeros
2. fill with mean value of feature
3. fill with last value using ffill()
4. interpolate linearly with interpolate()

```python
f, ax= plt.subplots(nrows=4, ncols=1, figsize=(20,15))

sns.lineplot(x=df['date'], y=df['drainage_volume'].fillna(0), ax=ax[0], color='darkorange', label='modified')
sns.lineplot(x=df['date'], y=df['drainage_volume'].fillna(np.inf), ax=ax[0], color='blue', label='original')
ax[0].set_title('fill with 0s')
ax[0].set_ylabel('drainage_volume')

sns.lineplot(x=df['date'], y=df['drainage_volume'].fillna(df['drainage_volume'].mean()), ax=ax[1], color='darkorange', label='modified')
sns.lineplot(x=df['date'], y=df['drainage_volume'].fillna(np.inf), ax=ax[1], color='blue', label='original')
ax[1].set_title('fill with mean')
ax[1].set_ylabel('drainage_volume')

sns.lineplot(x=df['date'], y=df['drainage_volume'].ffill(), ax=ax[2], color='darkorange', label='modified')
sns.lineplot(x=df['date'], y=df['drainage_volume'].fillna(np.inf), ax=ax[2], color='blue', label='original')
ax[2].set_title('fill with last value')
ax[2].set_ylabel('drainage_volume')

sns.lineplot(x=df['date'], y=df['drainage_volume'].interpolate(), ax=ax[3], color='darkorange', label='modified')
sns.lineplot(x=df['date'], y=df['drainage_volume'].fillna(np.inf), ax=ax[3], color='blue', label='original')
ax[3].set_title('fill with last value')
ax[3].set_ylabel('drainage_volume')

```

```python
f, ax= plt.subplots(nrows=4, ncols=1, figsize=(20,15))

sns.lineplot(x=df['date'], y=df['hydrometry'].fillna(0), ax=ax[0], color='darkorange', label='modified')
sns.lineplot(x=df['date'], y=df['hydrometry'].fillna(np.inf), ax=ax[0], color='blue', label='original')
ax[0].set_title('fill with 0s')
ax[0].set_ylabel('hydrometry')

sns.lineplot(x=df['date'], y=df['hydrometry'].fillna(df['hydrometry'].mean()), ax=ax[1], color='darkorange', label='modified')
sns.lineplot(x=df['date'], y=df['hydrometry'].fillna(np.inf), ax=ax[1], color='blue', label='original')
ax[1].set_title('fill with mean')
ax[1].set_ylabel('hydrometry')

sns.lineplot(x=df['date'], y=df['hydrometry'].ffill(), ax=ax[2], color='darkorange', label='modified')
sns.lineplot(x=df['date'], y=df['hydrometry'].fillna(np.inf), ax=ax[2], color='blue', label='original')
ax[2].set_title('fill with last value')
ax[2].set_ylabel('hydrometry')

sns.lineplot(x=df['date'], y=df['hydrometry'].interpolate(), ax=ax[3], color='darkorange', label='modified')
sns.lineplot(x=df['date'], y=df['hydrometry'].fillna(np.inf), ax=ax[3], color='blue', label='original')
ax[3].set_title('fill with last value')
ax[3].set_ylabel('hydrometry')
```

```python
df['drainage_volume'] = df['drainage_volume'].interpolate()
df['hydrometry'] = df['hydrometry'].interpolate()
df['depth'] = df['depth'].interpolate()
                           
```

downsample the data to weekly since we want a weekly forecast of the depth

```python
df.head()
```

```python
df_downsampled=df[['date','depth','temperature','drainage_volume','hydrometry']].resample('7D',on='date').mean().reset_index(drop=False)
df_downsampled.head()
```

```python
df_downsampled['rainfall']=df[['date', 'rainfall']].resample('7D',on='date').sum().reset_index(drop=False)[['rainfall']]
```

some feature engineering

```python
df2=df_downsampled.copy()
df2.dtypes
```

```python
df2['year']=pd.DatetimeIndex(df2['date']).year
df2['month']=pd.DatetimeIndex(df2['date']).month
df2['day']=pd.DatetimeIndex(df2['date']).day
df2['day_of_year'] = pd.DatetimeIndex(df2['date']).dayofyear
df2['quarter'] = pd.DatetimeIndex(df2['date']).quarter
```

```python
df2.head()
```

```python
month_in_year=12
df2['month_sin']=np.sin(2*np.pi*df2['month']/month_in_year)
df2['month_cos']=np.cos(2*np.pi*df2['month']/month_in_year)
```

time series decomposition

```python
df2.isna().sum()
```

```python
df2=df2.dropna(subset=['temperature', 'hydrometry'])
```

```python
df2.head()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decompose_cols=['rainfall', 'temperature', 'drainage_volume','hydrometry','depth']

for col in decompose_cols:
    decomp=seasonal_decompose(df2[col], period=52, model='additive', extrapolate_trend='freq')
    df2[f"{col}_trend"]=decomp.trend
    df2[f"{col}_seasonal"]=decomp.seasonal    
```
